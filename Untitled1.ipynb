```python
import pandas as pd
```

```python
file= open('./gangnam_time_zone111.txt', "r")
```

```python
lines = file.readlines()
```

```python
objs = []
for line in lines:
    print(line)
    cols = line.split('\t')
#     print(cols[1])
    obj = { "timestamp": cols[1], "user_id": cols[2], "longitude": cols[6], "latitude": cols[7], "camera_id": cols[3] }
    objs.append(obj)
```

```python
df = pd.DataFrame(objs)
```
